# Train a model with and without annotations and seed for reproducibility 

```python
reload(annotations_test)
annotated = annotations_test.run_test(True,seed=2)
non_annotated = annotations_test.run_test(False,seed=2)
```

```python
non_annotated.head(15)
```

```python
non_annotated = non_annotated.set_index(['feature', 'view', 'factor'])
# if swapping factors needed
non_annotated_tmp = non_annotated.unstack(level='factor')
non_annotated_tmp.head()
```

```python
tmp = np.copy(non_annotated_tmp.loc[:, 'inferred'].loc[:, 0].values)
non_annotated_tmp.loc[:, 'inferred'].loc[:, 0] = non_annotated_tmp.loc[:, 'inferred'].loc[:, 1].values
non_annotated_tmp.loc[:, 'inferred'].loc[:, 1] = tmp
non_annotated_tmp.head()
```

```python
non_annotated = non_annotated_tmp.stack(level='factor')
non_annotated = non_annotated.reset_index(['feature', 'view', 'factor'])
non_annotated.head()
```

```python
annotated['annotations'] = 'known'
non_annotated['annotations'] = 'unknown'
all_weights = pd.concat([annotated, non_annotated], axis =0)
all_weights.head()
```
